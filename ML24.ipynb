```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns                     #logistic regression/(Binary Classification) (Polynomial input)
```

```python
df=pd.read_csv(r"C:\\Users\\USER\\Downloads\\Desktop\\ML\\dataset\\bmi.csv")
df.drop("Gender",axis=1,inplace=True)
df.head()
```

```python
sns.scatterplot(x="Height", y="Weight", data=df, hue="Index")
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
X=pf.fit_transform(df[["Height","Weight"]])
pf.transform(df[["Height","Weight"]])
```

```python
x=pd.DataFrame(X)
x.head()
```

```python
y=df["Index"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test)
```
