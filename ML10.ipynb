```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\dataset\loan.csv")
df.head() 
```

```python
df.isnull().sum()
```

```python
df['ApplicantIncome'].fillna(df['ApplicantIncome'].mean(),inplace=True)
```

```python
sns.distplot(df["ApplicantIncome"])
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(df[["ApplicantIncome"]])
```

```python
df[["ApplicantIncome_ss"]]=pd.DataFrame(ss.transform(df[["ApplicantIncome"]]))
```

```python
plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
plt.title("Before Standardization")
sns.distplot(df["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("After Standardization")
sns.distplot(df["ApplicantIncome_ss"])


plt.show()
#data is sinked but nature remains same
```
