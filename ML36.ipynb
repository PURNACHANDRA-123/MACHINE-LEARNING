```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\Machine Learning\dataset\Placement_1.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
svr.fit(x_train,y_train)
```

```python
svr.score(x_test,y_test)*100
svr.score(x_train,y_train)*100
```

```python
sns.scatterplot(x='cgpa',y='package',data=df)
plt.plot(df['cgpa'],svr.predict(x),color='red')
plt.xlabel('cgpa')
```

```python
df1=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\Machine Learning\dataset\Position_Salaries.csv")
df1.drop('Position',axis=1,inplace=True)
df1.head()
```

```python
x1=df1[["Level"]]
y1=df1["Salary"]
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(x1,y1,test_size=0.2,random_state=41)
```

```python
from sklearn.svm import SVR
svr1=SVR(kernel='rbf')
svr1.fit(x_train1,y_train1)
svr1.score(x_test1,y_test1)
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Level',y='Salary',data=df1)
plt.plot(df1['Level'],svr1.predict(x1),color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
```
