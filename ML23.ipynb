```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns                     #logistic regression/(Binary Classification) (Polynomial input)
```

```python
df=pd.read_csv(r"C:\\Users\\USER\\Downloads\\Desktop\\ML\\dataset\\bmi.csv")
```

```python
df.drop("Gender",axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.scatterplot(x="Height", y="Weight", data=df, hue="Index")
```

```python
x=df[["Height","Weight"]]
y=df["Index"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LG=LogisticRegression()
LG.fit(x_train,y_train)
```

```python
LG.score(x_test,y_test)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=LG, legend=2)
```
