```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#one hot encoding for categorical data
#filing missing value from pd lib 
#or sklearn (good option storing in numerical form )
```

```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\dataset\loan.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Gender"].fillna(df["Gender"].mode()[0],inplace=True)
```

```python
df["Married"].fillna(df["Married"].mode()[0],inplace=True)
```

```python
en_df=df[["Gender","Married"]]
en_df
```

```python
pd.get_dummies(en_df)

#but this bool type we have convert into in int so using sklearn one hotencoding is good 
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop="first")  #droping 1st column after onehotencoding columns
```

```python
ar=ohe.fit_transform(en_df).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Married_No"])

```

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```
