```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
#logistic regression
```

```python
df=pd.read_csv(r"C:\\Users\\USER\\Downloads\\Desktop\\ML\\dataset\\Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.drop("User ID",axis=1,inplace=True)
```

```python
df.drop("Gender",axis=1,inplace=True)
```

```python
df.drop(columns="EstimatedSalary",inplace=True)
```

```python
df.head(),df.shape
```

```python
sns.scatterplot(x="Age",y="Purchased",data=df)
```

```python
x=df[["Age"]]
y=df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test)*100
```

```python
lg.predict([[30]])
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=df)
sns.lineplot(x="Age",y=lg.predict(x),data=df,c="r")
plt.show()
```
