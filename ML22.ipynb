```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns                     #binary classfication multiple inputs
```

```python
df=pd.read_csv(r"C:\\Users\\USER\\Downloads\\Desktop\\ML\\dataset\\placement-dataset.csv")
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=df,hue="placement")
```

```python
x=df.iloc[:,:-1]
y=df["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
model.predict([[4.6,120]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=model,legend=2)
```
