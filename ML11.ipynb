```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#feature scaling normalization
#min max scaling
```

```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\dataset\loan.csv")
df.head() 
```

```python
sns.displot(df["ApplicantIncome"],kde=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(df[["ApplicantIncome"]])
```

```python
df["ApplicantIncome minmax"]=ms.transform(df[["ApplicantIncome"]])
```

```python
df.head()
```

```python
plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
plt.title("before minmax scaling")
sns.distplot(df["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("after minmax scaling")
sns.distplot(df["ApplicantIncome minmax"])


plt.show()
```
