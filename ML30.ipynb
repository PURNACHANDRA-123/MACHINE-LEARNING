```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions
#naive bayes classifier
```

```python
df=pd.read_csv(r"C:\\Users\\USER\\Downloads\\Desktop\\ML\\dataset\\placement-dataset.csv")
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head(),df.isnull().sum()
```

```python
x=df.iloc[:,0:-1]
y=df["placement"]
```

```python
sns.scatterplot(x="cgpa",y="iq",hue="placement",data=df)
```

```python
sns.kdeplot(data=df["cgpa"])
```

```python
sns.kdeplot(x="cgpa",y="iq",hue="placement",data=df)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.xlabel("cgpa")
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.xlabel("cgpa")
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.xlabel("cgpa")
```
