```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt     
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv(r"C:\\Users\\USER\\Downloads\\Desktop\\ML\\Machine Learning\\dataset\\placement-dataset.csv")
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head(),df.isnull().sum(),df.describe()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=df,hue="placement")
plt.show()
```

```python
x=df.drop('placement',axis=1)
y=df['placement']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(x_train,y_train) 
print(model.score(x_test,y_test)*100)
print(model.score(x_train,y_train)*100)
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=model, legend=2)
plt.xlabel('cgpa')
plt.ylabel('iq')
plt.show()
```
