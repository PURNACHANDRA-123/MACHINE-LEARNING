```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#linear regression
#y=mx+c m means slope and c means intercept (c is the y-intercept of the line. It represents the value of y when x is equal to 0.)
```

```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\dataset\Placement_1.csv")
```

```python
df.head()
```

```python
x=df[["cgpa"]]
y=df[["package"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_prdict=lr.predict(x)
```

```python
sns.scatterplot(x="cgpa",y="package",data=df,label="data")
plt.plot(df["cgpa"],y_prdict,color="red",label="regression line")
plt.legend(["org","regression line"])
plt.savefig("regression.png")
plt.show()
```
