```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#outliers removal using zscore
```

```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Desktop\ML\dataset\loan.csv")
df.head() 
```

```python
df.describe()
```

```python
#Z score
#z=(x-mean)/std where x is the value

```

```python
sns.boxplot(x=df['ApplicantIncome'])
```

```python
sns.displot(df['ApplicantIncome'], kde=True)
```

```python
min_app=df['ApplicantIncome'].mean()-3*df['ApplicantIncome'].std()
max_app=df['ApplicantIncome'].mean()+3*df['ApplicantIncome'].std()
```

```python
min_app,max_app
```

```python
new_df=df[df["ApplicantIncome"]<=max_app]
new_df
```

```python
sns.boxplot(x=new_df['ApplicantIncome'])
```

```python
z_score=(df["ApplicantIncome"]-df["ApplicantIncome"].mean())/(df["ApplicantIncome"].std())
z_score
```

```python
df["z_score"]=z_score
df
```

```python
df[df["z_score"]<3]
```
